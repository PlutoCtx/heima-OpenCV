```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('./image/yangzi.jpg')
```

```python
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
# 实例化检测器
face_cas = cv.CascadeClassifier('D:\\Program Files\\OpenCV\\opencv\\sources\\data\\haarcascades\\haarcascade_frontalface_default.xml')
face_cas.load("D:\\Program Files\\OpenCV\\opencv\\sources\\data\\haarcascades\\haarcascade_frontalface_default.xml")

eyes_cas = cv.CascadeClassifier("D:\\Program Files\\OpenCV\\opencv\\sources\\data\\haarcascades\\haarcascade_eye.xml")
eyes_cas.load("D:\\Program Files\\OpenCV\\opencv\\sources\\data\\haarcascades\\haarcascade_eye.xml")

# 导入有问题请参考https://blog.csdn.net/LeungSr/article/details/119985008
```

```python
# 人脸检测
face_rects = face_cas.detectMultiScale(gray,scaleFactor =1.2,minNeighbors = 3,minSize=(32,32))
```

```python
# 绘制人脸检测眼睛
for facerect in face_rects:
    x,y,w,h = facerect
    cv.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
    roi_color = img[y:y+h,x:x+w]
    roi_gray = gray[y:y+h,x:x+w]
    eyes = eyes_cas.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),3)
```
