```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt
```

```python
dogsp = cv.imread("./image/dogsp.jpeg")
```

```python
doggause = cv.imread("./image/dogGauss.jpeg")
```

# 1 均值滤波

```python
dog = cv.blur(dogsp, (5,5))
```

```python
plt.imshow(dog[:, :, ::-1])
```

# 2 高斯滤波

```python
dog = cv.GaussianBlur(doggause, (3, 3), 1)
```

```python
plt.imshow(dog[:, :, ::-1])
```

# 中值滤波

```python
dog = cv.medianBlur(dogsp, 3)
```

```python
plt.imshow(dog[:, :, ::-1])
```
