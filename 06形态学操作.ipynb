```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt
```

# 腐蚀与膨胀

```python
img = cv.imread("./image/letter.png")
```

```python
plt.imshow(img[:, :, ::-1])
```

```python
# 创建核结构
kenel = np.ones([5,5], np.uint8)
```

```python
# 腐蚀
img2 = cv.erode(img, kenel)
```

```python
plt.imshow(img2[:, :, ::-1])
```

```python
# 膨胀
img1 = cv.dilate(img, kenel)
```

```python
plt.imshow(img1[:, :, ::-1])
```
